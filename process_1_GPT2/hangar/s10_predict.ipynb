## 0- `Config`

```python
from transformers import GPT2Tokenizer, GPT2Model
import torch
```

## 1- `Text to Predict`

```python
# Example text format for prediction
text_to_predict = "Unknown " + "JCHAIN"  # 'Unknown' acts as a placeholder for Cell_Type 
```

## 2- `Preprocess Text`

```python

tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
tokenizer.pad_token = tokenizer.eos_token

model = GPT2Model.from_pretrained('gpt2')

inputs = tokenizer(text_to_predict, return_tensors="pt", padding=True, truncation=True, max_length=512)

with torch.no_grad():
    outputs = model(**inputs)
    embedding = outputs.last_hidden_state.mean(dim=1)
```

## 3- `Prediction`

```python
index_SVM = 0 
SVM_All_Model = SVM_All_Models[index_SVM]

prediction = SVM_All_Model.predict(embedding.numpy())

print("Predicted Cell Type:", prediction[0]) 
SVM_All_Model
```
