# Combinations 

```python
from itertools import product    
```

```python
# Outlier
options_outlier = ["Free_v0",
                   "ZScore_v0","ZScore_v1",
                   "ISO_v0","ISO_v1",
                   "LOF_v0","LOF_v1", 
                   ] 
```

```python
# Autoencoder
options_autoencoder = [ "Free_v0",
                        "Basic_v0","Basic_v1","Basic_v2",
                        "Denoising_v0","Denoising_v1","Denoising_v2","Denoising_v3",    
                        "Dropout_v0","Dropout_v1","Dropout_v2","Dropout_v3", 
                        "Combined_v0","Combined_v1","Combined_v2","Combined_v3","Combined_v4","Combined_v5","Combined_v6","Combined_v7",
                        ]
```

```python
# Dimension Reductions 
# PCA 
n_components_set = [2, 3, 50, 51] 

options_dr_pca = [] 
for i in n_components_set:
    options_dr_pca.append(f"PCA_{i}") 

# TSNE 
n_components_set    = [2, 3]
perplexity_set      = [5, 30, 50] 
learning_rate_set   = [3, 10, 100, 500] 

options_dr_tsne = []
combinations_tsne = list(product(n_components_set, perplexity_set, learning_rate_set)) 
for a,b,c in combinations_tsne:
    options_dr_tsne.append(f"TSNE_nc{a}_p{b}_lr{c}")   

# UMAP 
options_dr_umap = []    
n_components_set  = [2, 3] 
n_neighbors_set   = [15, 50, 60] 
min_dist_set      = [0.1, 0.5, 0.99] 
combinations_umap = list(product(n_components_set, n_neighbors_set, min_dist_set))
for a,b,c in combinations_umap:
    options_dr_umap.append(f"UMAP_nc{a}_nn{b}_md{c}")  

options_dr = options_dr_pca + options_dr_tsne + options_dr_umap 
```

```python
combinations = list(product(options_outlier, options_autoencoder, options_dr)) 
print("Lenght of combinations: ", len(combinations)) 
```
