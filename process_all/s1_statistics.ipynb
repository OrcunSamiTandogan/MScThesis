```python
subject = "ZScore_v0_ZScore_v0_Basic_v1_Basic_v1_StandardScaler_StandardScaler_PCA_n3_Free"
subject = "Free_v0_Free_v0_Free_TSNE_nc3_p50_lr10"

def partialize(subject):
    if isinstance(subject, str):
        partial = subject.split("_")
        try: 
            outlier_1 = partial[0] + "_" + partial[1] 
            outlier_2 = partial[2] + "_" + partial[3] 
            autoencoder_1 = partial[4] + "_" + partial[5] 
            autoencoder_2 = partial[6] + "_" + partial[7] 
            normalizer_1 = partial[8] 
            normalizer_2 = partial[9] 
            dimension_reduction_1 = partial[10] + "_" + partial[11] 
            dimension_reduction_2 = "_".join(partial[12:])
            return outlier_1, outlier_2, autoencoder_1, autoencoder_2, normalizer_1, normalizer_2, dimension_reduction_1, dimension_reduction_2
        except IndexError:
            # Handle cases where the split list is shorter than expected
            return (partial[0] + "_" + partial[1] if len(partial) > 1 else "Missing",
                    "Missing",
                    partial[2] + "_" + partial[3] if len(partial) > 3 else "Missing",
                    "Missing",
                    partial[4] if len(partial) > 4 else "Missing",
                    "Missing",
                    partial[5] + "_" + partial[6] if len(partial) > 6 else "Missing",
                    "Missing")
    else:
        # Handle non-string inputs
        return "Missing", "Missing", "Missing", "Missing", "Missing", "Missing", "Missing", "Missing"

outlier_1, outlier_2, autoencoder_1, autoencoder_2, normalizer_1, normalizer_2, dimension_reduction_1, dimension_reduction_2 = partialize(subject)
print("Outlier_1:", outlier_1)
print("Outlier_1:", outlier_2)
print("Autoencoder 1:", autoencoder_1)
print("Autoencoder 2:", autoencoder_2)
print("Normalizer 1:", normalizer_1)
print("Normalizer 2:", normalizer_2)
print("Dimension Reduction 1:", dimension_reduction_1)
print("Dimension Reduction 2:", dimension_reduction_2) 
```

```python
Subject_Model_Parsed_Sided_Subset = Subject_Model[["Text_Embedding", "Applications",]]

Subject_Model_Parsed_Sided_Subset[['Outlier_1', 'Outlier_2', 'Autoencoder_1', 'Autoencoder_2', 'Normalizer_1', 'Normalizer_2', 'Dimension_Reduction_1', 'Dimension_Reduction_2']] = Subject_Model_Parsed_Sided_Subset.apply(
    lambda row: pd.Series(partialize(row['Applications'])), axis=1) 

Subject_Model_Parsed_Sided_Subset_Drop = Subject_Model_Parsed_Sided_Subset.drop(columns=["Text_Embedding","Applications"]) 

Subject_Model_Important = Subject_Model[["Text_Embedding", "Model", "Accuracy", "Precision", "F1", "Recall", "Parameters", "Process"]] 
Subject_Model_Final = pd.concat([Subject_Model_Important, Subject_Model_Parsed_Sided_Subset_Drop], axis = 1) 
```
